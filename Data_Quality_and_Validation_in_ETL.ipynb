Question 1: Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

Data quality refers to how accurate, complete, consistent, and reliable the data is throughout the ETL process. Good quality data helps organizations make correct business decisions.

It is more than just data cleaning because it also includes validating data, checking for duplicates, maintaining consistency, following business rules, and ensuring that the data is loaded correctly into the target system. Data cleaning is only one part of maintaining overall data quality.


Question 2: Explain why poor data quality leads to misleading dashboards and incorrect decisions.

Dashboards and reports are only as good as the data used to create them. If the data contains errors, missing values, or duplicate records, the reports will also be incorrect. This can lead managers to make poor business decisions, such as ordering the wrong amount of inventory or targeting the wrong customers. Therefore, maintaining high data quality is essential for accurate reporting and analysis.

Question 3: What is duplicate data? Explain three causes in ETL pipelines.

Duplicate data means the same information appears more than once in a dataset when it should only appear once.

Some common causes are:

Data is loaded multiple times because of ETL job failures or repeated imports.
The same data is collected from different systems without proper duplicate checking.
Human errors during data entry, such as entering the same transaction more than once.

Removing duplicate records helps improve data quality and prevents incorrect analysis.

Question 4: Differentiate between exact, partial, and fuzzy duplicates.

Exact duplicates are records that are completely identical in every field.

Partial duplicates are records where most of the information is the same, but one or two fields are different.

Fuzzy duplicates are records that refer to the same person or item but have small spelling or formatting differences.

For example, "Rahul Mehta" and "Rahul M." may refer to the same customer, making them fuzzy duplicates.

Question 5: Why should data validation be performed during transformation rather than after loading?

Data validation should be done during the transformation stage because it allows errors to be identified and corrected before the data reaches the final database. This prevents invalid or incomplete data from entering the system and reduces the time needed for corrections later. Validating data early also improves the overall quality of the ETL process.

Question 6: Explain how business rules help in validating data accuracy. Give an example.

Business rules are conditions that data must satisfy before it is accepted into the system. These rules help ensure that the data is accurate, complete, and consistent.

For example, if a company's rule says that the transaction amount must always be greater than zero, then any record with a negative or zero amount will be rejected. This helps maintain reliable and accurate data.

Question 7: Write an SQL query on Sales_Transactions to list all duplicate keys and their counts using the business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount).

Answer

- SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS Duplicate_Count
FROM Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING COUNT(*) > 1;
